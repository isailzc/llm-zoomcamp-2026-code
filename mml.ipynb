```python
import numpy as np
import matplotlib.pyplot as plt

# =========================
# Malla para los planos
# =========================
x1 = np.linspace(-2, 4, 30)
x2 = np.linspace(-2, 4, 30)

X1, X2 = np.meshgrid(x1, x2)

# =========================
# Planos
# =========================

# (1) x1 + x2 + x3 = 3
Z1 = 3 - X1 - X2

# (2) x1 - x2 + 2x3 = 2
Z2 = (2 - X1 + X2) / 2

# (3) 2x1 + 3x3 = 5
# Este plano no depende de x2
Z3 = (5 - 2 * X1) / 3

# =========================
# Recta de intersección
# =========================

a = np.linspace(-2, 4, 200)

x1_line = 5/2 - 3*a/2
x2_line = 1/2 + a/2
x3_line = a

# =========================
# Gráfica
# =========================

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Planos
ax.plot_surface(
    X1, X2, Z1,
    alpha=0.4,
    label='Plano 1'
)

ax.plot_surface(
    X1, X2, Z2,
    alpha=0.4,
    label='Plano 2'
)

ax.plot_surface(
    X1, X2, Z3,
    alpha=0.4,
    label='Plano 3'
)

# Recta de intersección
ax.plot(
    x1_line,
    x2_line,
    x3_line,
    linewidth=4,
    label='Intersección'
)

# Etiquetas
ax.set_xlabel('$x_1$')
ax.set_ylabel('$x_2$')
ax.set_zlabel('$x_3$')

ax.set_title('Intersección de tres planos')

# Ajustar ángulo de vista
ax.view_init(elev=25, azim=45)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 1. Configurar la figura en 3D
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# 2. Crear una malla (grid) para los ejes X y Y (o X1 y X2)
# Usaremos un rango para x1 y a (que es x3) para despejar x2 cuando sea necesario,
# o simplemente una malla estándar x1, x2 para despejar x3.
x1_range = np.linspace(0, 5, 20)
x2_range = np.linspace(-2, 4, 20)
X1, X2 = np.meshgrid(x1_range, x2_range)

# Despejamos x3 (Z) de cada ecuación para graficar los planos:
# (1) x3 = 3 - x1 - x2
Z1 = 3 - X1 - X2

# (2) x3 = (2 - x1 + x2) / 2
Z2 = (2 - X1 + X2) / 2

# (3) 2*x1 + 3*x3 = 5 -> x3 = (5 - 2*x1) / 3
# Para que se grafique como plano en la malla, Z3 no depende de X2
Z3 = (5 - 2 * X1) / 3

# Graficar los planos con cierta transparencia (alpha)
ax.plot_surface(X1, X2, Z1, alpha=0.5, color='cyan', label='Plano 1: x1+x2+x3=3')
ax.plot_surface(X1, X2, Z2, alpha=0.5, color='magenta', label='Plano 2: x1-x2+2x3=2')
ax.plot_surface(X1, X2, Z3, alpha=0.4, color='yellow', label='Plano 3: 2x1+3x3=5')

# 3. Graficar la línea de intersección usando la parametrización
# Tu parametrización: x1 = 5/2 - 3/2*a, x2 = 1/2 + 1/2*a, x3 = a
a = np.linspace(-1, 3, 100)  # Rango del parámetro 'a'
line_x1 = 2.5 - 1.5 * a
line_x2 = 0.5 + 0.5 * a
line_x3 = a

# Dibujar la línea de intersección (más gruesa y de color rojo)
ax.plot(line_x1, line_x2, line_x3, color='red', linewidth=3, label='Línea de Intersección')

# 4. Configuraciones estéticas del gráfico
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('X3')
ax.set_title('Intersección de Planos en $\mathbb{R}^3$')

# Truco para mostrar la leyenda correctamente con superficies 3D
# Matplotlib a veces tiene problemas duplicando leyendas de superficies.
from matplotlib.lines import Line2D
legend_elements = [
    Line2D([0], [0], color='cyan', lw=4, alpha=0.5, label='$x_1 + x_2 + x_3 = 3$'),
    Line2D([0], [0], color='magenta', lw=4, alpha=0.5, label='$x_1 - x_2 + 2x_3 = 2$'),
    Line2D([0], [0], color='yellow', lw=4, alpha=0.5, label='$2x_1 + 3x_3 = 5$'),
    Line2D([0], [0], color='red', lw=3, label='Línea paramétrica')
]
ax.legend(handles=legend_elements)

# Ajustar los límites de los ejes para centrar la vista en la intersección
ax.set_xlim(0, 5)
ax.set_ylim(-2, 4)
ax.set_zlim(-1, 3)

plt.show()
```
